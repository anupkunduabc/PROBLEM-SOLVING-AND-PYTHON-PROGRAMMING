```python
# Python String Operations Demonstration

text1 = "Hello"
text2 = "World"
text3 = "   Python Programming   "

print("Initial strings:")
print("text1 =", repr(text1))
print(text1)
print("text2 =", repr(text2))
print("text3 =", repr(text3))

# 1. Basic Operations
print("\n--- Basic Operations ---")
print("Concatenation (text1 + text2):", text1 + text2)
print("Repetition (text1 * 3):", text1 * 3)
print("Length of text1 (len):", len(text1))

# 2. Indexing & Slicing
print("\n--- Indexing & Slicing ---")
print("First character of text1:", text1[0])
print("Last character of text1:", text1[-1])
print("Slice text1[1:4]:", text1[1:4])
print("Reverse text1:", text1[::-1])

# 3. Common String Methods (Case Conversion)
print("\n--- Case Conversion ---")
print("Upper case:", text1.upper())
print("Lower case:", text1.lower())
print("Title case:", text3.title())
print("Capitalize:", text2.capitalize())
print("Swapcase:", text1.swapcase())

# 4. Whitespace Removal
print("\n--- Whitespace Removal ---")
print("Original text3:", repr(text3))
print("strip():", repr(text3.strip()))
print("lstrip():", repr(text3.lstrip()))
print("rstrip():", repr(text3.rstrip()))

# 5. Searching & Checking
print("\n--- Searching & Checking ---")
print("'Py' in text3:", "Py" in text3)
print("'Java' not in text3:", "Java" not in text3)
print("text3.find('Python'):", text3.find("Python"))
print("text3.count('m'):", text3.count("m"))
print("text1.startswith('He'):", text1.startswith("He"))
print("text2.endswith('ld'):", text2.endswith("ld"))

# 6. Replace & Split/Join
print("\n--- Replace, Split & Join ---")
print("Replace 'World' with 'Python':", text2.replace("World", "Python"))
words = text3.strip().split()
print("Split text3 into words:", words)
print("Join words with '-':", "-".join(words))

# 7. String Formatting
print("\n--- String Formatting ---")
name = "Alice"
age = 22
print("Using format(): Name: {}, Age: {}".format(name, age))
print("Using f-string: Name: {0}, Age: {1}".format(name, age))
print(f"Using f-string directly: Name: {name}, Age: {age}")

# 8. Miscellaneous
print("\n--- Miscellaneous ---")
print("isalnum() on 'Hello123':", "Hello123".isalnum())
print("isalpha() on 'Hello':", "Hello".isalpha())
print("isdigit() on '123':", "123".isdigit())
print("isspace() on '   ':", "   ".isspace())
print("text3.center(30, '-'):", text3.strip().center(30, '-'))

```
